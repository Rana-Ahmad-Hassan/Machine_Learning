Linear Regression

Linear regression is a statistical method used to model the relationship between a dependent variable (target variable) and one or more independent variables (predictor variables) by fitting a linear equation to the observed data. The goal is to create a model that can predict the value of the dependent variable based on the values of the independent variables.

Equation:

Y = β0 + β1X + ε

Where:

- Y: Dependent variable (target variable)
- X: Independent variable (predictor variable)
- β0: Intercept or constant term
- β1: Slope coefficient
- ε: Error term

Types of Linear Regression:

1. Simple Linear Regression (SLR): One independent variable is used to predict the dependent variable.
2. Multiple Linear Regression (MLR): More than one independent variable is used to predict the dependent variable.
3. Multivariate Linear Regression: Multiple dependent variables are predicted using one or more independent variables.

These types of linear regression are widely used in various fields, including economics, finance, marketing, and social sciences, to analyze and predict relationships between variables.

`Simple Linear Regression`

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#load the dataset
data = pd.read_csv("tips.csv")
print(data.head())
```

```python
sns.scatterplot(data=data, x='total_bill', y='tip')
plt.show()

sns.histplot(data=data['total_bill'], kde=True)
plt.show()

mean = np.mean(data['total_bill'])
print(mean)
```

```python
#Seperate the feature and the output

X = data[['total_bill']]
Y = data['tip']
```

```python
#lets split the data into test and train
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
#train the model
model = LinearRegression()
model.fit(X_train, Y_train)
```

```python
#take the intercept and the slop
print(model.intercept_)
print(model.coef_)
```

```python
#lets prrdict
y_pred = model.predict(X_test)
y_pred
```

```python
#evaluate the error
print(f'MSE = {mean_squared_error(Y_test, y_pred)}')
print(f'R2 = {r2_score(Y_test, y_pred)}')
print(f'RMSE = {np.sqrt(mean_squared_error(Y_test, y_pred))}')
```
