`SVM (Support Vector Machine)
`
SVM is a supervised machine learning algorithm used for classification and regression tasks. It aims to find the best hyperplane (decision boundary) that separates the data into different classes with the maximum margin.

`Key Concepts:
`
1. Hyperplane: A line or plane that separates the data into different classes.
2. Margin: The distance between the hyperplane and the nearest data points (support vectors).
3. Support Vectors: Data points that lie closest to the hyperplane and define the margin.

`How SVM Works:
`
1. Data Preparation: Prepare the dataset with labeled examples.
2. Kernel Selection: Choose a suitable kernel function to transform the data into a higher-dimensional space (if necessary).
3. Hyperplane Selection: Find the optimal hyperplane that maximizes the margin between classes.
4. Classification: Use the trained SVM model to classify new, unseen data points.

`Types of SVM:
`
1. Linear SVM: Uses a linear kernel function for classification.
2. Non-Linear SVM: Uses non-linear kernel functions (e.g., polynomial, radial basis function (RBF)) for classification.

`Advantages:
`
1. High Accuracy: SVM can achieve high accuracy in classification tasks.
2. Robustness: SVM is robust to noise and outliers in the data.
3. Flexibility: SVM can handle high-dimensional data and non-linear relationships.

`Applications:
`
1. Image Classification: SVM is used in image classification tasks, such as object detection and recognition.
2. Text Classification: SVM is used in text classification tasks, such as spam detection and sentiment analysis.
3. Bioinformatics: SVM is used in bioinformatics for tasks like protein classification and gene expression analysis.

SVM is a powerful algorithm for classification and regression tasks, and its applications are diverse and widespread.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import train_test_split
```

```python
data = sns.load_dataset('iris')
data.head()
```

```python
X = data.drop(columns='species', axis=1)
Y= data['species']
```

```python
train_X, test_X, train_Y, test_Y = train_test_split(X, Y, train_size=0.20, random_state=42)
```

```python
model = SVC()
model.fit(train_X, train_Y)
```

```python
y_pred = model.predict(test_X)
y_pred
```

```python
# ðŸ“Š Evaluate performance
print("\nConfusion Matrix:\n", confusion_matrix(test_Y, y_pred))
print("\nClassification Report:\n", classification_report(test_Y, y_pred))
```

```python
sns.heatmap(data=confusion_matrix(test_Y, y_pred), annot=True)
plt.show()
```
